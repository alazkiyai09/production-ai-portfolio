## 3. Fraud Analysis Demo

````python
# Simulated fraud analysis
fraud_report = {
    "reporter": "Bank ABC",
    "transaction_id": "TXN-12345",
    "amount": 15000,
    "description": "Unusual large transaction at odd hours",
    "evidence": ["Location mismatch", "High amount", "Unusual timing"]
}

print("ğŸ” Fraud Analysis\n")
print("=" * 60)
print(f"Report: {fraud_report['reporter']}")
print(f"Transaction: {fraud_report['transaction_id']}")
print(f"Amount: ${fraud_report['amount']:,}")
print(f"Evidence: {', '.join(fraud_report['evidence'])}")

# Calculate risk
risk_score = min(
    (50 if fraud_report['amount'] > 10000 else 0) +
    (30 if 'Location mismatch' in fraud_report['evidence'] else 0) +
    (20 if 'Unusual timing' in fraud_report['description'].lower() else 0),
    100
)
print(f"\nğŸ¯ Risk Score: {risk_score}/100")

if risk_score > 70:
    print("ğŸš¨ HIGH RISK - Immediate escalation required")
elif risk_score > 40:
    print("âš ï¸  MEDIUM RISK - Further investigation needed")
else:
    print("âœ… LOW RISK - Normal processing")
````
