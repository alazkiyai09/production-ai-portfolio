## 3. Risk Scoring Demo

````python
def calculate_risk_score(transaction: dict, patterns: list) -> float:
    """Calculate fraud risk score (0-100)."""
    risk = 0
    
    # Amount risk
    if transaction.get('amount', 0) > 10000:
        risk += 30
    elif transaction.get('amount', 0) > 5000:
        risk += 15
    
    # Frequency risk
    if transaction.get('frequency', 'low') == 'high':
        risk += 40
    
    # Location risk
    if transaction.get('location_mismatch', False):
        risk += 25
    
    return min(risk, 100)

# Test risk scoring
test_transactions = [
    {"amount": 15000, "frequency": "low", "location_mismatch": False},
    {"amount": 2000, "frequency": "high", "location_mismatch": True},
    {"amount": 8000, "frequency": "medium", "location_mismatch": False}
]

print("ðŸŽ¯ Risk Scoring\n")
for i, txn in enumerate(test_transactions, 1):
    score = calculate_risk_score(txn, fraud_patterns)
    status = "ðŸš¨ HIGH RISK" if score > 50 else "âš ï¸ MEDIUM" if score > 25 else "âœ… LOW RISK"
    print(f"Txn {i}: {status} (Score: {score}/100)")
    print(f"  Amount: ${txn['amount']:,} | Frequency: {txn['frequency']} | Location Mismatch: {txn['location_mismatch']}")
    print()
````
